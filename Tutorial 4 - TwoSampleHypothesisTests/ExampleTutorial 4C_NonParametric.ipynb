# Non-Parametric Tests

A group of researchers in the field of biomaterials and tissue engineering is investigating the potential of a bioactive hydrogel for promoting bone tissue regeneration. They want to evaluate if their bioactive hydrogel promotes faster bone healing compared to a commonly used synthetic scaffold material. To assess this, they conducted a study using 22 bone defect models in animal subjects, measuring the time taken in days for initial bone regeneration as the primary outcome. Nine of the bone defect models were treated with the bioactive hydrogel (Group 1), while the remaining 14 models were treated with the synthetic scaffold material (Group 2).

__Importing relevant libraries!__

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats
from scipy import stats  
from scipy.stats import norm
import seaborn as sns
import statsmodels.api as sm
```

## Part 1
What type of statistical test should we do here and why? Utilize Q-Q Plots and the Shapiro-wilk test to evaluate this.

```python
#Creating the Dataset
hydrogel_data = np.array([10.7,19.1,10.2,10.9,15.8,12.3,10.2,17.1,10.7])
hydogel_n = len(hydrogel_data)

syn_data= np.array([16.4,17.2,16.8,17.1,16.9,17.0,16.2,15.0,17.9,20.2,17.2,17.3,16.0,19.1])
syn_n = len(syn_data)

#Generating a box-plot
plt.figure(1)
box_plot = sns.boxplot(data=[hydrogel_data, syn_data], palette=['red', 'gold'])
box_plot.set(ylabel='Healing time (days)')
box_plot.set_xticklabels(['Hydrogel', 'Synthetic'])
plt.savefig('P4_Boxplot.png', dpi=300, bbox_inches='tight')
```

__Conduct the Shapiro-Wilk Test__

```python
#Shapiro-Wilk Test
print('\n')
print("Hydrogel:",stats.shapiro(hydrogel_data))
print("Synthetic:",stats.shapiro(syn_data))
```

__Generate Q-Q Plots__

```python
#Generating QQ Plots
plt.figure(2)
sm.qqplot((np.sort(hydrogel_data)),line='45',loc=np.mean(hydrogel_data),scale=np.std(hydrogel_data,ddof=1))
plt.title('QQ plot of the Hydrogel')
plt.savefig('P4_QQ_hydrogel.png', dpi=300, bbox_inches='tight')

plt.figure(3)
sm.qqplot((np.sort(syn_data)),line='45',loc=np.mean(syn_data),scale=np.std(syn_data,ddof=1))
plt.title('QQ plot of the Synthetic')
plt.savefig('P4_QQ_Syn.png', dpi=300, bbox_inches='tight')
```
